```python
%matplotlib inline
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
sys.path.append('../spectra/')
import spectra
```

```python
S = spectra.Spectra('../test_data/xrd.txt')
plt.plot(S.x,S.y,'r-')
```

```python
S.reset()
S.find_background(cutoff=1/4500, order=3)
plt.plot(S.x,S.y,'r-',S.x,S.bg,'b-')
```

```python
S.subtract_background()
plt.plot(S.x,S.y,'r-')
S.find_peaks()
plt.plot(S.x,S.y,'b-',S.x[S.peak_pos],S.y[S.peak_pos],'oy')
```

```python
S.build_model()
S.fit_data()
```

```python
plt.plot(S.x,S.y,'g-',S.x,S.out.init_fit,'b-', S.x,S.out.best_fit,'k-')
```
